Prevendo o Preço de Fechamento de um Ativo
Em um mundo de dados financeiros, uma equipe de analistas que queria entender o comportamento do mercado e, mais importante ainda, prever o futuro do resultado de um ativo no final do dia. Eles tinham uma base de dados rica, contendo informações sobre o desempenho diário de um ativo, como a Abertura, Máxima, Mínima, Variação Percentual (Var%) e, é claro, o valor de fechamento do ativo, chamado de Último.

O Desafio: Prever o Preço de Fechamento (Último)
O objetivo da equipe era simples, mas desafiador: prever o preço de fechamento de um ativo no final de cada dia. Para isso, era preciso construir um modelo preditivo que utilizasse as variáveis anteriores, como a Abertura do dia, o preço máximo atingido, o preço mínimo e a variação do mercado ao longo do dia. Eles sabiam que a tendência do tempo (a evolução do preço com o passar dos dias) era crucial para fazer boas previsões.

Preparando os Dados: A Magia da Limpeza
O primeiro passo da jornada foi preparar os dados. A equipe tinha os dados em mãos, mas eles estavam desorganizados. A coluna Data, por exemplo, estava em formato de texto, e para fazer um bom trabalho de previsão, ela precisava ser convertida para um formato que o modelo pudesse entender. O primeiro feitiço foi usar a função pd.to_datetime() para transformar a coluna de datas em objetos de data do pandas. Assim, eles conseguiram trabalhar com a cronologia corretamente.

Além disso, a coluna de Var%, que representava a variação percentual, estava com símbolos de porcentagem e vírgulas. Nada que um bom str.replace() não resolvesse! A equipe limpou os dados, transformando-os para o formato adequado para análise, removendo os sinais de porcentagem e substituindo vírgulas por pontos.

Dividindo os Dados: Treinamento e Teste
Agora que os dados estavam preparados, era hora de construir o modelo de previsão. A equipe sabia que, para evitar overfitting (ajuste excessivo ao conjunto de dados), eles precisavam dividir os dados em duas partes: uma para treinar o modelo e outra para testá-lo.

O código usou a função train_test_split() do sklearn para separar os dados em dois conjuntos. 70% dos dados seriam usados para treinar o modelo, e 30% seriam usados para testar como o modelo se comportaria em dados não vistos. Eles também escolheram não embaralhar os dados com shuffle=False para manter a sequência temporal intacta – afinal, os dados financeiros dependem da ordem cronológica dos eventos.

A Mágica da Regressão Linear
Agora, a parte mais importante da história: o modelo! A equipe decidiu usar o clássico modelo de Regressão Linear. A razão era simples: eles queriam entender a relação linear entre as variáveis de entrada (Abertura, Máxima, Mínima, Var%) e a variável de saída (Último).

A equipe treinou o modelo com os dados de treinamento usando o comando modelo.fit(X_train, y_train), onde X_train eram as características do ativo e y_train era o preço de fechamento real. O modelo começou a entender como as variáveis influenciavam o preço de fechamento.

Fazendo Previsões: O Futuro Revelado
Com o modelo treinado, a equipe estava pronta para fazer previsões. Usando os dados de teste (X_test), o modelo foi capaz de prever os valores de Último para as datas em que eles não tinham os valores reais. Eles usaram a função modelo.predict(X_test) para gerar essas previsões.

Avaliação: O Poder do R²
Como saber se o modelo estava realmente bom? A equipe usou uma métrica chamada R² (coeficiente de determinação) para avaliar o desempenho do modelo. O R² indica quanto da variação dos dados pode ser explicada pelo modelo. Um valor de R² próximo de 1 significa que o modelo é ótimo, enquanto um valor próximo de 0 sugere que o modelo não tem poder de previsão.

Com o código r2_score(y_test, previsoes), eles calcularam a acurácia do modelo e descobriram o quão bem ele estava prevendo o preço de fechamento.

Visualizando o Futuro: O Gráfico
Agora que tinham suas previsões, era hora de visualizar os resultados. A equipe decidiu usar um gráfico de linhas para mostrar como as previsões (linhas "Previsão") se comparam com os valores reais de fechamento (linhas "Real").

O gráfico gerado pelo código mostrava claramente a evolução do preço de fechamento ao longo do tempo, permitindo à equipe ver como as previsões se comportavam em relação ao que realmente aconteceu no mercado. O gráfico também ajudava a identificar padrões e possíveis pontos de melhoria para o modelo.

O Resultado: O Modelo Está Pronto
Ao final, a equipe teve um modelo funcional de previsão, capaz de prever o resultado de fechamento de um ativo com base em variáveis do dia, como Abertura, Máxima, Mínima e Var%. Embora o modelo fosse baseado em uma abordagem simples de Regressão Linear, ele já fornecia boas previsões e um ponto de partida sólido para futuras melhorias.

```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    confusion_matrix,
)

# Importação das bibliotecas necessárias (já incluídas no código original)

from google.colab import drive
```

```python
import pandas as pd

try:
  # Tenta carregar o arquivo do Google Drive
  df = pd.read_csv('/content/drive/My Drive/Dados Históricos - Ibovespa (10).txt', sep=',')
except FileNotFoundError:
  print("Arquivo não encontrado no Google Drive. Tentando baixar do GitHub...")
  try:
    # Tenta baixar o arquivo do GitHub
    url = 'https://raw.githubusercontent.com/SandersonSB/Tecchallenge2/refs/heads/main/Dados%20IBOVESPA' # Substitua pelo link correto do seu arquivo no GitHub
    df = pd.read_csv(url, sep=',')
  except Exception as e:
    print(f"Erro ao carregar o arquivo do GitHub: {e}")
    df = None  # Define df como None em caso de erro

if df is not None:
  print("Dados carregados com sucesso!")
  # Aqui você pode continuar com o processamento do DataFrame
  print(df.head()) # Mostra as primeiras linhas do dataframe para verificação
else:
  print("Não foi possível carregar os dados.")
```

```python
# Convertendo a coluna 'Data' para o formato datetime
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')

# Definindo a coluna 'Data' como índice do DataFrame
df = df.set_index('Data')

# Convertendo a coluna 'Var%' para numérico
# Substituindo vírgula por ponto e removendo o sinal de porcentagem
df['Var%'] = df['Var%'].str.replace(',', '.').str.rstrip('%').astype(float)

# Criando as features (X) e o target (y)
X = df[['Abertura', 'Máxima', 'Mínima', 'Var%']]
y = df['Último']

# Dividindo os dados em conjuntos de treinamento e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42, shuffle=False)

# Importando o modelo de regressão linear
from sklearn.linear_model import LinearRegression

# Criando e treinando o modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
previsoes = modelo.predict(X_test)

# Avaliando o modelo (usando R² como métrica)
from sklearn.metrics import r2_score
r2 = r2_score(y_test, previsoes)
print(f"Acuracidade: {r2}")

# Plotando o gráfico das previsões vs valores reais
plt.figure(figsize=(20, 2))
plt.plot(y_test.index, y_test.values, label='Real')
plt.plot(y_test.index, previsoes, label='Previsão')
plt.xlabel('Data')
plt.ylabel('Último')
plt.title('Previsão do Último vs Valor Real')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: preciso que com essa mesma logica seja criado um que veja a digerença da linha real e da linha de precisão e me traga as extremidades, que  seja exibido em grafico de bolhas a maior diferença e menor fiferença

# Calcula a diferença entre os valores reais e as previsões
diferencas = abs(y_test - previsoes)

# Cria um DataFrame com as datas, valores reais, previsões e diferenças
df_diferencas = pd.DataFrame({'Data': y_test.index, 'Real': y_test.values, 'Previsão': previsoes, 'Diferença': diferencas})

# Encontra a maior e menor diferença
maior_diferenca = df_diferencas['Diferença'].max()
menor_diferenca = df_diferencas['Diferença'].min()

# Encontra as datas correspondentes à maior e menor diferença
data_maior_diferenca = df_diferencas.loc[df_diferencas['Diferença'] == maior_diferenca, 'Data'].iloc[0]
data_menor_diferenca = df_diferencas.loc[df_diferencas['Diferença'] == menor_diferenca, 'Data'].iloc[0]

# Plota o gráfico de bolhas
plt.figure(figsize=(10, 6))
plt.scatter(df_diferencas['Data'], df_diferencas['Diferença'], s=df_diferencas['Diferença'] * 10, alpha=0.5)  # Ajuste o tamanho das bolhas conforme necessário

# Destaca os pontos da maior e menor diferença
plt.scatter(data_maior_diferenca, maior_diferenca, color='red', s=maior_diferenca * 10, label=f'Maior Diferença: {maior_diferenca:.2f} ({data_maior_diferenca.strftime("%d/%m/%Y")})')
plt.scatter(data_menor_diferenca, menor_diferenca, color='green', s=menor_diferenca * 10, label=f'Menor Diferença: {menor_diferenca:.2f} ({data_menor_diferenca.strftime("%d/%m/%Y")})')

plt.xlabel('Data')
plt.ylabel('Diferença (|Real - Previsão|)')
plt.title('Diferença entre Valores Reais e Previsões com Destaque para as Extremidades')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True)
plt.tight_layout()  # Ajusta o layout para evitar que as labels se sobreponham
plt.show()
```
